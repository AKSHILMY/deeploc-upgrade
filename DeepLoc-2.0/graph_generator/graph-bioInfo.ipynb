```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
file_path = 'bio-info.csv'
data = pd.read_csv(file_path)
```

```python
# Check the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Extract the column names
locations = data.iloc[:, 0]  # Protein location names
models = data.columns[1:]    # Model names
```

```python
# Set the number of locations and models
num_locations = len(locations)
num_models = len(models)
print(f'Number of locations: {num_locations} \nNumber of models: {num_models}')
```

```python
# Create an array of indices for the locations with gaps
gap = 0.0001  # This sets the gap between groups of bars
ind = np.arange(num_locations) * (num_models + gap)
```

```python
# Set the width of each bar
width = 0.8

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(15, 8))

# Plot each model's performance as a separate set of bars
for i, model in enumerate(models):
    ax.bar(ind + i * width, data[model], width, label=model)

# Add labels, title, and legend
ax.set_xlabel('Protein Locations')
ax.set_ylabel('Performance')
ax.set_title('Model Performance by Protein Location')
ax.set_xticks(ind + width * (num_models - 1) / 2)
ax.set_xticklabels(locations, rotation=90)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
