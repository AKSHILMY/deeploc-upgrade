```python
import scipy.io
import pandas as pd
```

```python
mat = scipy.io.loadmat('./dataset_4802.mat') 
mat = {k:v for k, v in mat.items() if k[0] != '_'}
```

```python
Y_4802_df = pd.DataFrame(mat['Y_4802'])
label_name_flat = [item[0][0] for item in mat['Y_4802_label']]
label_name_df = pd.DataFrame(label_name_flat, columns=['label_name'])
# protein_list_flat = [item[0] for item in mat['protein_list'][0]]
# protein_list_df = pd.DataFrame(protein_list_flat, columns=['ACC'])
sequence_4802_flat = [item[0][0] for item in mat['Sequence']]
sequence_4802_df = pd.DataFrame(sequence_4802_flat, columns=['Sequence'])
```

```python
Y_4802_df.columns = [label_name_df['label_name']]
Y_4802_df['sequence'] = sequence_4802_df['Sequence']
# Y_4802_df['ACC'] = protein_list_df['ACC']
```

```python
default = pd.read_csv("../multisub_5_partitions_unique.csv")
Y_4802_df = pd.read_csv("./Y_4802.csv")
```

```python
default.columns,Y_4802_df.columns
```

```python
# Manually removing columns
columns_to_remove_in_Y_4802 = ['label_name',
'Apical Plasma Membrane',
'Basolateral Plasma Membrane',
'Cellular Component Unknown',
'Centrosome',
# 'Cytoplasm',
'Cytoplasmic Vesicles',
'Cytoskeleton',
'ERGIC',
'Early Endosomes',
# 'Endoplasmic Reticulum',
'Endosomes',
# 'Extracellular',
# 'Golgi Apparatus',
'Golgi Cis Cisterna',
'Golgi Trans Cisterna',
'Golgi Trans Face',
'Inner Mitochondrial Membrane',
'Late Endosomes',
'Lipid Particles',
# 'Lysosomes',
'Medial-Golgi',
'Melanosome',
'Microtubule',
'Microtubule Organizing Center',
# 'Mitochondria',
'Nuclear Envelope',
'Nucleolus',
# 'Nucleus',
'Outer Mitochondrial Membrane',
# 'Peroxisome',
# 'Plasma Membrane',
'Sarcolemma',
'Secretory Granule',
'Secretory Vesicles',
'Synaptic Vesicles',
'Tight Junction',
'Transport Vesicle',
# 'sequence'
]
Y_4802_df = Y_4802_df.drop(columns_to_remove_in_Y_4802, axis=1)
```

```python
columns_to_rename_in_Y_4802 = {
    # old name : new name
    'Endoplasmic Reticulum' : 'Endoplasmic reticulum',
    'Golgi Apparatus' : 'Golgi apparatus',
    'Lysosomes' : 'Lysosome/Vacuole',
    'Plasma Membrane' : 'Cell membrane',
    'Mitochondria' : 'Mitochondrion',
    'sequence' : 'Sequence',
}

Y_4802_df = Y_4802_df.rename(columns=columns_to_rename_in_Y_4802)
```

```python
columns_to_add_in_Y_4802 = {
    # column name : default value
    'Kingdom' : '',
    'Partition' : '',
    'Membrane' : 0.0,
    'Plastid' : 0.0,
}
for col in columns_to_add_in_Y_4802:
    Y_4802_df[col] = columns_to_add_in_Y_4802[col]
```

```python
combined_df = pd.concat([default,Y_4802_df],ignore_index=True)
```
